**Scatter Chart**

A scatter chart is best when you want to show the relationship between two numerical parameters. In your data, the "Count" column is numerical, so it should be one axis. The other axis can be a numerical transformation, such as grouping by "Cause category" or "Outcome of Incident" (converted to counts). For example, you can compare “Total number of Accidents” vs. “Total Injured” across different cities. This helps show whether more accidents always lead to more injuries.

```python
# Filter dataset to only include rows with 'Total number of Accidents' and 'Total Injured'
accidents = df[df['Outcome of Incident'] == 'Total number of Accidents']
injured = df[df['Outcome of Incident'] == 'Total Injured']

# Merge on city and cause to align values
scatter_data = accidents.merge(
    injured,
    on=['Million Plus Cities', 'Cause category', 'Cause Subcategory'],
    suffixes=('_Accidents', '_Injured')
)

# Plot scatter chart
plt.figure(figsize=(8, 6))
plt.scatter(
    scatter_data['Count_Accidents'],
    scatter_data['Count_Injured'],
    alpha=0.6,
    color='blue'
)

# Add labels and title
plt.title('Scatter Chart: Accidents vs Injured')
plt.xlabel('Total Number of Accidents')
plt.ylabel('Total Injured')

# Show chart
plt.show()
```

The scatter chart shows how the total number of accidents relates to the total number of injured. If the points go upward as accidents increase, it means more accidents usually lead to more injuries. Points that stand out show unusual cases where the number of injuries does not match the number of accidents.

**Bubble Chart**

A bubble chart adds a third dimension by using bubble size. Here, the x-axis and y-axis can still be numerical values like “Total number of Accidents” and “Total Injured”, while the bubble size can represent “Persons Killed” or “Minor Injury” counts. You can also use “Million Plus Cities” or “Cause category” as grouping labels to distinguish bubbles by color.

```python
# Filter dataset for needed outcomes
accidents = df[df['Outcome of Incident'] == 'Total number of Accidents']
injured = df[df['Outcome of Incident'] == 'Total Injured']
killed = df[df['Outcome of Incident'] == 'Persons Killed']

# Merge data on city and cause to align values
bubble_data = accidents.merge(
    injured,
    on=['Million Plus Cities', 'Cause category', 'Cause Subcategory'],
    suffixes=('_Accidents', '_Injured')
).merge(
    killed,
    on=['Million Plus Cities', 'Cause category', 'Cause Subcategory']
)

# Plot bubble chart
plt.figure(figsize=(10, 7))
plt.scatter(
    bubble_data['Count_Accidents'],
    bubble_data['Count_Injured'],
    s=bubble_data['Count'],   # bubble size = persons killed
    alpha=0.5,
    c='orange',
    edgecolors='black'
)

# Add labels and title
plt.title('Bubble Chart: Accidents vs Injured (Bubble Size = Persons Killed)')
plt.xlabel('Total Number of Accidents')
plt.ylabel('Total Injured')

# Show chart
plt.show()
```

The bubble chart shows the link between the total number of accidents (x-axis) and the total number of injured (y-axis), while the bubble size represents the number of persons killed. Bigger bubbles mean more deaths in that category or city. This makes it easy to see not only how accidents relate to injuries but also where the impact is more severe in terms of fatalities.

**Line Chart**
A line chart is useful for showing trends over time. It connects data points with lines, making it easy to see increases, decreases, or patterns across a timeline or sequence.

A line chart is effective for showing changes in road accidents or injuries across a period. For example, plotting the “Total number of Accidents” by months or years can reveal whether cases are rising, falling, or staying steady. This helps in spotting patterns and understanding the overall trend rather than just looking at single values.

```python
# Filter only 'Total number of Accidents'
accidents = df[df['Outcome of Incident'] == 'Total number of Accidents']

# Group by City and sum counts
city_trend = accidents.groupby('Million Plus Cities')['Count'].sum()

# Plot line chart
plt.figure(figsize=(12, 6))
plt.plot(city_trend.index, city_trend.values, marker='o', linestyle='-', color='blue')

# Add labels and title
plt.title('Total Number of Accidents by City')
plt.xlabel('Million Plus Cities')
plt.ylabel('Total Number of Accidents')

# Rotate x-axis labels for readability
plt.xticks(rotation=90)

# Show chart
plt.show()
```

**Column Chart**
A column chart uses vertical bars to compare values across categories. It is effective when you want to highlight differences in size or frequency between groups.

```python
# Filter only 'Total number of Accidents'
accidents = df[df['Outcome of Incident'] == 'Total number of Accidents']

# Group by City and sum counts
city_counts = accidents.groupby('Million Plus Cities')['Count'].sum()

# Plot column chart (vertical bars)
plt.figure(figsize=(12, 6))
city_counts.plot(kind='bar', color='skyblue', edgecolor='black')

# Add labels and title
plt.title('Total Number of Accidents by City')
plt.xlabel('Million Plus Cities')
plt.ylabel('Total Number of Accidents')

# Rotate x-axis labels for readability
plt.xticks(rotation=90)

# Show chart
plt.show()
```

**BAR CHART**

We aim to identify the top 10 happiest countries in 2024 and compare their happiness scores. This analysis will highlight which nations rank the highest and allow us to observe how their scores differ, providing insights into global well-being distribution.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("World-happiness-report-2024.csv")

# Define correct columns
country_col = "Country name"
score_col = "Ladder score"

# Sort by Happiness Score (descending) and select Top 10
top10 = df.sort_values(by=score_col, ascending=False).head(10)

# Create Vertical Bar Chart
plt.figure(figsize=(10,6))
plt.bar(top10[country_col], top10[score_col], color="skyblue", edgecolor="black")

# Labels and Title
plt.xlabel("Country")
plt.ylabel("Happiness Score (Ladder Score)")
plt.title("Top 10 Happiest Countries (2024)")

# Rotate x-axis labels for readability
plt.xticks(rotation=45, ha="right")

# Show chart
plt.show()

```

This chart uses vertical bars to represent the Top 10 happiest countries. Each bar’s height corresponds to the country’s happiness score, making it easy to compare across nations. Vertical bar charts are widely used and intuitive, which makes them suitable for presentations and reports.

**SCATTER PLOT**

We aim to examine the relationship between a country’s economic performance (Log GDP per capita) and its Happiness Score (Ladder score) in 2024. The goal is to determine whether wealthier nations are generally happier, or if other factors might also play a role.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("World-happiness-report-2024.csv")

# Define correct columns
x_col = "Log GDP per capita"
y_col = "Ladder score"

# Create Scatter Plot
plt.figure(figsize=(10,6))
plt.scatter(df[x_col], df[y_col], color="skyblue", edgecolor="black", alpha=0.7)

# Labels and Title
plt.xlabel("Log GDP per Capita")
plt.ylabel("Happiness Score (Ladder Score)")
plt.title("Relationship Between GDP per Capita and Happiness (2024)")

# Add grid for clarity
plt.grid(True, linestyle="--", alpha=0.6)

plt.show()
```

A scatter plot is ideal because it shows the relationship between two continuous variables. In this chart, each point represents a country, with its GDP per capita on the x-axis and happiness score on the y-axis.

The plot helps us see overall trends: if points rise as GDP increases, this suggests a positive correlation between wealth and happiness. Outliers also become visible (countries with relatively high happiness despite lower GDP). Adding this scatter plot complements the bar chart by moving beyond rankings and exploring possible causal relationships.

**LINE CHART**

We want to compare the average happiness scores across world regions in 2024. This helps identify which regions report higher well-being and which regions face challenges in terms of happiness.

```python
region_scores = df.groupby("Regional indicator")["Ladder score"].mean().reset_index()

# Sort regions by score for better visualization
region_scores = region_scores.sort_values(by="Ladder score", ascending=False)

# Create Line Chart
plt.figure(figsize=(12,6))
plt.plot(region_scores["Regional indicator"], region_scores["Ladder score"], marker="o", linestyle="-", color="skyblue")

# Labels and Title
plt.xlabel("Region")
plt.ylabel("Average Happiness Score (Ladder Score)")
plt.title("Average Happiness Score by Region (2024)")
plt.xticks(rotation=45, ha="right")

# Add grid
plt.grid(True, linestyle="--", alpha=0.6)

plt.show()
```

The line chart is used here to emphasize regional comparisons of average happiness. While bar charts are also common for this type of data, a line chart highlights trends and makes it easy to see the rise and fall of scores across different regions.

By plotting each region in order, the visualization provides a clear picture of how happiness differs geographically, helping policymakers and researchers identify regions that are thriving and those that may require more attention.

**BUBBLE CHART**

We aim to analyze the relationship between GDP per capita, happiness, and life expectancy across countries in 2024, while also distinguishing patterns across different regions.

```python
x_col = "Log GDP per capita"
y_col = "Ladder score"
size_col = "Healthy life expectancy"
color_col = "Regional indicator"

# Normalize bubble sizes for better visualization
sizes = df[size_col] * 50

# Create Bubble Chart
plt.figure(figsize=(12,8))
scatter = plt.scatter(
    df[x_col],
    df[y_col],
    s=sizes,
    alpha=0.6,
    c=pd.factorize(df[color_col])[0],
    cmap="tab10",
    edgecolor="black"
)

# Labels and Title
plt.xlabel("Log GDP per Capita")
plt.ylabel("Happiness Score (Ladder Score)")
plt.title("Bubble Chart of Happiness, GDP, and Life Expectancy (2024)")

# Legend for regions
plt.legend(*scatter.legend_elements(), title="Region", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

A bubble chart is chosen because it can represent three variables simultaneously. In this visualization:

1.The x-axis shows GDP per capita (wealth).

2.The y-axis shows Happiness Score.

3.The bubble size reflects Healthy Life Expectancy.

4.The bubble color differentiates regions.

This allows us to observe not only whether wealthier countries tend to be happier, but also how life expectancy contributes to well-being. Regional color coding further enhances understanding by revealing clusters —for example, European countries may group differently from African or Asian ones.
The bubble chart is particularly useful for identifying outliers (countries with unusual patterns) and spotting general trends among multiple variables.

**PIE CHART**

We aim to visualize how average happiness scores are distributed across different regions of the world in 2024. This helps identify which regions contribute more to global well-being.

```python
region_scores = df.groupby("Regional indicator")["Ladder score"].mean().reset_index()

# Create Pie Chart
plt.figure(figsize=(8,8))
plt.pie(
    region_scores["Ladder score"],
    labels=region_scores["Regional indicator"],
    autopct="%1.1f%%",
    startangle=140,
    colors=plt.cm.tab10.colors
)

plt.title("Share of Average Happiness Score by Region (2024)")
plt.show()
```
